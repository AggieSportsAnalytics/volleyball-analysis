```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.multioutput import MultiOutputClassifier
from sklearn.metrics import classification_report
import pickle
```

```python
df = pd.read_csv('merged-data (1).csv')
```

```python
df.drop(['Attend'], axis=1, inplace=True)
```

```python
feature_cols = [col for col in df.columns if 'team' in col.lower()]
target_cols = ['Sets', 'Kills','Assists', 'Digs', 'Block Solos']
```

```python
for col in target_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
for col in target_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df['Errors']
```

```python
features = df[feature_cols]
target = df['Errors']
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target)
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
with open('flask-server/models/error-model.pkl', 'wb') as f:
    pickle.dump(rf, f)
```
